### Express duration of songs in seconds
(Originally, the duration is expressed in terms of milliseconds)

```python
print("Before:", tracks[:1][["duration_ms"]])
tracks["duration"] = tracks["duration_ms"].apply(lambda x : (x/1000))        # apply()
print("After:", tracks[:1][["duration"]])
```

### Sample out a dataset of size 0.4% of original tracks dataset and create a regression plot between loudness and energy

```python
sample_data = tracks.sample(int(0.004*len(tracks)))
len(sample_data)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y="loudness",x="energy",color="c").set(title="Loudness vs. Energy correlation")
```

#### Looking at the above plot is can be said that Loudness increases with increase in Energy

### Regression plot between popularity and acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y="popularity",x="acousticness",color="b").set(title="Popularity vs. Acousticness correlation")
```

#### Looking at the above plot is can be said that Popularity decreases with increase in Acousticness

## Display number of songs released per year since 1922 in form of distribution plot

```python
# Typecasting release dates from str to date-time objects
#tracks["dates"]=tracks.index.get_level_values("release_date")
tracks.release_date=pd.to_datetime(tracks.release_date)
years=tracks.release_date.dt.year
```

##### pip install --user seaborn==0.11.0

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year since 1922")
```

## Show variation of duration of songs over the years

```python
total_duration=tracks.duration
fig,ax=plt.subplots(figsize=(18,7))
fig=sns.barplot(x=years,y=total_duration,ax=ax,errwidth=False).set(title="Duration vs. Year")
plt.xticks(rotation=90)
plt.show()
```

```python
total_duration=tracks.duration
sns.set_style(style="whitegrid")
fig,ax=plt.subplots(figsize=(10,5))
fig=sns.lineplot(x=years,y=total_duration,ax=ax).set(title="Duration vs. Year")
plt.xticks(rotation=60)
plt.show()
```

# Analysing SpotifyFeatures.csv

```python
features.head()
```

### Show duration of songs of different genres

```python
plt.title("Duration of songs of different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre",x="duration_ms",data=features)
plt.xlabel("Duration in ms")
plt.ylabel("Genre")
plt.show()
```

### Top 5 genres according to popularity

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=features.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="Top 5 genres according to popularity")
plt.show()
```
